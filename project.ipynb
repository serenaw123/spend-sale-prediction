# Histogram:

```python
heights=pd.DataFrame({'Height':[160,162,165,167,167,167,168,170,170,171,172,173,175,178,180,182,183,185,187,207]})

mean_height=heights['Height'].mean()
median_height=heights['Height'].median()
mode_height=heights['Height'].mode()[0]

plt.figure(figsize=(8,6))
plt.hist(heights['Height'], bins=15,color='blue', edgecolor='black',alpha=0.7)

plt.axvline(mean_height, color='red',linestyle='dashed',linewidth=2, label=f'Mean: {mean_height:.1f}cm')
plt.axvline(median_height, color='green',linestyle='dashed',linewidth=2, label=f'Median: {median_height:.1f}cm')
plt.axvline(mode_height, color='orange',linestyle='dashed',linewidth=2, label=f'Mode: {mode_height:.1f}cm')

plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Heights in a Small Town")
plt.show()
```
